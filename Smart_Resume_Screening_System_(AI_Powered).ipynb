```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
job_description = "We need an AI/ML developer with expertise in Python, Machine Learning, and TensorFlow."

resumes = [
    "I am a Python developer skilled in Machine Learning and Data Science.",
    "I have experience in Java development and project management.",
    "Expert in TensorFlow, Python, and Deep Learning models."
]
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
all_text = [job_description] + resumes
tfidf_matrix = vectorizer.fit_transform(all_text)
```

```python
jd_vector = tfidf_matrix[0]
resume_vectors = tfidf_matrix[1:]
scores = cosine_similarity(jd_vector, resume_vectors).flatten()
```

```python
df_results = pd.DataFrame({
    'Resume': resumes,
    'Match_Score': scores * 100
})
```
